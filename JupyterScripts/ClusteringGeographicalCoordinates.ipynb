```python
import pandas as pd 
import numpy as np
import hdbscan
import pickle
import os

from IPython.display import Image
os.chdir("D:\\GoogleDrive\\InterviewAssignment\\GoJekDriverAcceptanceGit\\GoJekAssignment")


train  = pd.read_csv("data\\processed\\dataset.csv")
train = train.iloc[:5000]

df_cord_driver = train[["driver_latitude", "driver_longitude"]]
clusterer_driver = hdbscan.HDBSCAN(min_cluster_size=2000, metric='haversine',prediction_data = True)
driver_cluster_labels = clusterer_driver.fit_predict(np.radians(df_cord_driver))
pickle.dump(clusterer_driver, open('models\\clusterer_driver_small.pkl', 'wb'))

df_cord_customer = train[["pickup_latitude", "pickup_longitude"]]
clusterer_customer = hdbscan.HDBSCAN(min_cluster_size=2000, metric='haversine',prediction_data = True)
customer_cluster_labels = clusterer_customer.fit_predict(np.radians(df_cord_customer))
pickle.dump(clusterer_customer, open('models\\clusterer_customer_small.pkl', 'wb'))
```
